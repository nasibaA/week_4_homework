```python
import pandas as pd
import numpy as np
file_to_load = "purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
print(purchase_data.head())
```

```python
#player Count.
total_players = len (purchase_data["SN"].value_counts())
players = pd.DataFrame({"total players":[total_players]})
players
```

```python
#Purchasing Analysis (Total)

unique_items = purchase_data["Item ID"].nunique()
average_price = purchase_data["Price"].mean()
purchase_count = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

my_df = pd.DataFrame({"Number of unique items": [unique_items],"Average purchase price": [average_price],"Total Number of Purchases": [purchase_count],"Total Revenue":[total_revenue]})

my_df

```

```python
#Gender Demographics
gender_count_1 = purchase_data["Gender"].value_counts()
gender_count_2 = pd.DataFrame(gender_count_1)
gender_count_2['Percentage of Players'] = 100 * gender_count_2 / gender_count_2.sum()
gender_count_2.rename(columns={'Gender':'Total Count'}, inplace=True)
gender_count_2 = gender_count_2.round(decimals = 2)
gender_count_2
```

```python
#Purchasing Analysis (Gender)
purchase_count = purchase_data["Purchase ID"].count()

average_price = purchase_data["Price"].mean()

total_revenue = purchase_data["Price"].sum()

per_person = total_revenue/purchase_count

Gender_Demographics = pd.DataFrame({"Purchase Count":[purchase_count],"Average Price":[average_price],"Total Revenue":[total_revenue],"Per Person":[per_person]})      

Gender_Demographics.index.name = ("Gender")
Gender_Demographics
```

```python
# Age Demographics
age_bins = [0,9,14,19,24,29,34,39,99]
bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39",">40"]
    
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels=bin_names)
purchase_data    
grouped_by_age = purchase_data.groupby("Age Group")   
count_age = grouped_by_age["SN"].nunique()
average_price_by_age = grouped_by_age["Price"].mean()
total_revenue_by_age = grouped_by_age["Price"].sum()
per_person = total_revenue_by_age/count_age
  
Age_demographics = pd.DataFrame({"Purchase Count":count_age,"Average Purchase Price":average_price_by_age,"Total Purchase Value":total_revenue_by_age,"Average Purchase Total Per Person":per_person})    
Age_demographics  
    
    
```

```python
#Top Spenders
spenders = purchase_data.groupby("SN")
spenders_count = spenders["Item ID"].count()
spenders_mean = spenders["Price"].mean()
spender_sum = spenders["Price"].sum()
spenders_1 = pd.DataFrame({"Purchase Count":spenders_count,"Average Purchase Price":spenders_mean,"Total Purchase Value":spender_sum})
spenders_2 = spenders_1.sort_values(["Total Purchase Value"], ascending=False).head()

spenders_2
```

```python
# Most Popular Items
most_popular = purchase_data.groupby(["Item ID","Item Name","Price"])
most_popular
most_popular_1 = most_popular["Purchase ID"].count()
sum_price = most_popular["Price"].sum()
item_price= sum_price/most_popular_1
items = pd.DataFrame({"Purchase Count":most_popular_1,"Item Price":item_price,"Total Purchase Value":sum_price})
popular = items.sort_values(["Purchase Count"],ascending=False).head()
popular
```

```python
#Most Profitable Items
profitable_items = purchase_data.groupby(["Item ID","Item Name","Price"])
profitable_items
profitable_item_count = profitable_items["Purchase ID"].count()
profitable_item_sum = profitable_items["Price"].sum()
sub_profitable = profitable_item_sum/profitable_item_count
profitable_item = pd.DataFrame({"Purchase Count":profitable_item_count,"Total Purchase Value":profitable_item_sum})

most_profitable_item = profitable_item.sort_values(["Purchase Count"],ascending=False).head()

most_profitable_item




```
